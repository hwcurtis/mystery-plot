```python
import pandas as pd
# Pots survey over 3,4 years 

# Create a DataFrame with the demographic data
data = {
	'Age Range': ['11-32'],
	'Female Percentage': [85],
	'Ratio': [5.9],
	'White Patients': [98],
	'Black or Asian Patients Percentage': [1],
	'Hispanic Patients Percentage': [3.3]
}
# Average of people affected over a 3-4 year span

df = pd.DataFrame(data)
print(df)
import matplotlib.pyplot as plt

# Plotting a bar chart for the demographic data
plt.figure(figsize=(8, 6))
# Exclude 'Age Range' as it is not numeric
numeric_data = {k: v for k, v in data.items() if k != 'Age Range'}
plt.bar(numeric_data.keys(), [v[0] for v in numeric_data.values()], color='skyblue')
plt.xlabel('Demographic Categories')
plt.ylabel('Values')
plt.title('Demographic Data of People Suffering from POTS')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Data for symptoms comparison between females and males
symptoms_data = {
	"Symptom": ["Palpitations", "Nausea", "Lightheadedness and Brain Fog"],
	"Female Percentage": [78, 83, 90],  # Ensure all values are provided
	"Male Percentage": [48, 57, 65]     # Ensure all values are provided
}

# Create a DataFrame for the symptoms data
symptoms_df = pd.DataFrame(symptoms_data)
print(symptoms_df)

# Plotting the symptoms data
plt.figure(figsize=(8, 6))
x = range(len(symptoms_data["Symptom"]))  # Use symptoms_data instead of undefined POTS_data
# Adjust x positions for side-by-side bars
x_female = [i - 0.2 for i in x]  # Shift Female bars slightly to the left
x_male = [i + 0.2 for i in x]    # Shift Male bars slightly to the right

# Plot the bars
plt.bar(x_female, symptoms_data["Female Percentage"], width=0.4, label="Female", color="pink")
plt.bar(x_male, symptoms_data["Male Percentage"], width=0.4, label="Male", color="blue")

# Set x-axis labels
plt.xticks(x, symptoms_data["Symptom"], rotation=45)
plt.xlabel("Symptoms")
plt.ylabel("Percentage")
plt.title("Comparison of POTS  Symptoms Between Females and Males")
plt.legend()
plt.tight_layout()
plt.show()
```
