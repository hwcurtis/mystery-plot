```python
import pandas as pd

# Data for the mystery
data = {
    "Time": ["8:pM", "8:30 PM", "9:00 PM", "9:30 PM"],
    "Event": ["Dinner starts", "lights go out", "Murder occurs", "Detective arrives"],
    "Suspect": ["None", "All", "Unknown", "All"],
    "Clue": ["Knife missing", "Scream heard", "Blood found", "Footprints discovered"]
}

# Create a DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

# Plotting the timeline
plt.figure(figsize=(10, 6))
plt.plot(df["Time"], df["Event"], marker="o", linestyle="-", color="red")

# Add labels
plt.title("Mystery Murder Timeline", fontsize=16)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Event", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
