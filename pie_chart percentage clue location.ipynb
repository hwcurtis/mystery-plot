```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a pie chart
clues = pd.DataFrame({
        'Clue': ['Bloody knife', 'Wet Clothes', 'Witness statement'],
        'Found_At': ['Boatdoc', 'New Boat', 'Paper found in the cabin']
    })

# Plot the pie chart
clues['Clue'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6, 6))
plt.title('Distribution of Clues')
plt.ylabel('')
plt.show()

# Define the data for the second pie chart
data = {
    'Location': ['Boatdoc', 'New Boat', 'Paper found in the cabin'],
    'Votes': [50, 30, 20]
}

df = pd.DataFrame(data)

# Calculate percentages
df['Percentage'] = (df['Votes'] / df['Votes'].sum()) * 100

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Percentage'], labels=df['Location'], autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Votes on Where Items Would Be Found')
plt.show()
```
